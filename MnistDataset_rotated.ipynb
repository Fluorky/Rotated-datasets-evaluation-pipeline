```python
import tensorflow as tf
import matplotlib.pyplot as plt

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

print(f"Training images shape: {train_images.shape}")  # (60000, 28, 28)
print(f"Training labels shape: {train_labels.shape}")  # (60000,)
```

```python

plt.figure(figsize=(10, 10))

for i in range(9):
    plt.subplot(1, 9, i + 1)
    plt.imshow(train_images[i], cmap='gray')
    plt.title(f"Label: {train_labels[i]}")
    plt.axis('off')

plt.show()
```

```python
from PIL import Image
import numpy as np
```

```python
rotated_images = np.array([np.array(Image.fromarray(image).rotate(30)) for image in train_images])
```
